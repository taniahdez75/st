```python
# fix the ? and the datatype problem in the first 3 columns only
for i in range(0,4):
  X[i] = X[i].str.replace('?','nan').values.astype(float)
print(X.head(15))
```

```python
# impute the mean of the cols that have NaN
import numpy as np
# Impute the 'NaN'  with mean of the values
for i in range(0,1557):
  X[i] = X[i].fillna(X[i].mean())
print(X.head(15))
```

```python
# Scale the data
from sklearn import preprocessing
minmaxScaler = preprocessing.MinMaxScaler()
X_tran = pd.DataFrame(minmaxScaler.fit_transform(X))
X_tran.head()
```

```python
# Creating a high dimension dataset
X_hd = pd.DataFrame(pd.np.tile(X, (1, 20)))
```

```python
# Printing the dimension of the data set
X_hd.shape
```

```python
from sklearn.model_selection import train_test_split
# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_hd, Y, test_size=0.3, random_state=123)
```

```python
# try to fit this data to logistic regression model to predict whether the image is an ad or no ad
from sklearn.linear_model import LogisticRegression
import time
# Defining the LogisticRegression function
benchmarkModel = LogisticRegression(max_iter=10000)
# Starting a timing function
t0=time.time()
# Fitting the model
benchmarkModel.fit(X_train, y_train)
# Finding the end time 

print("Total training time:", round(time.time()-t0, 3), "s")
```

```python
# Predicting using the model
pred = benchmarkModel.predict(X_test)
print('Accuracy of Logistic regression model prediction on test set: {:.2f}'.format(benchmarkModel.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
# Confusion Matrix for the model
print(confusion_matrix(y_test, pred))
# Classification report for the model
print(classification_report(y_test, pred))
```
